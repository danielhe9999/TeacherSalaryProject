```python
totalfinal = df['Total Final Salary:'].tolist()
insurance = df['Insurance:'].tolist()
benefits = df['Mandatory Benefits:'].tolist()
experience = df['Certificated Experience:'].tolist()
```

```python
newtotalfinal = []
for item in totalfinal:
    item = item.replace('$', '')
    item = item.replace(',', '')
    item = int(item)
    newtotalfinal.append(item)

newinsurance = []
for item in insurance:
    item = item.replace('$', '')
    item = item.replace(',', '')
    item = int(item)
    newinsurance.append(item)
    
newbenefits = []
for item in benefits:
    item = item.replace('$', '')
    item = item.replace(',', '')
    item = int(item)
    newbenefits.append(item)

```

```python
import numpy as np
newtotalfinalNP = np.array(newtotalfinal)
newinsuranceNP = np.array(newinsurance)
newbenefitsNP = np.array(newbenefits)
```

```python
newinsuranceNP
```

```python
tot = np.add(newtotalfinalNP, newinsuranceNP)
```

```python
total = np.add(tot, newbenefitsNP)
```

```python
df['Sum of all'] = total
```

```python
experience
```

```python
newexperience = []
for row in experience:
    if (row == 0):
        newexperience.append(0)
    else:
        row = row.replace('years', '')
        row = float(row)
        newexperience.append(row)
```

```python
schoolnames = df['School Name:'].tolist()
```

```python
high = 0
middle = 0
elem = 0
for school in schoolnames:
    if "High School" in school:
        high = high + 1
    elif "Middle School" in school:
        middle = middle +1
    elif "Junior High" in school:
         middle = middle +1
    elif "Intermediate" in school:
        middle = middle +1
    elif "Elementary" in school:
        elem = elem + 1
    elif "Elem" in school:
        elem = elem + 1
```

```python
#Linear Regression
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt 

X = finalframe.drop(['Sum of all', 'Age'], axis = 1)
y = finalframe['Sum of all']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25) 
  
regr = LinearRegression() 
regr.fit(X_train, y_train) 
print(regr.score(X_test, y_test)) 
```
