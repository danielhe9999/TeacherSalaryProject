```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
```

```python
URL = 'http://data.kitsapsun.com/projects/wa-school/'
page = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
counties = soup.find_all('div', class_="filter_target") #38 results2 filter targets, 38 counties
```

```python
districtlinks = []
districtnames = []

for i in range (len(counties)):
    for a in counties[i].find_all('a', href=True):
        districtlinks.append(a['href'])
        districtnames.append(a.text)
```

```python
schoollinks = []
schoolnames = []

for link in districtlinks:
    Districtpage = requests.get(link, headers={'User-Agent': 'Mozilla/5.0'})
    Districtsoup = BeautifulSoup(Districtpage.content, 'html.parser')
    listofSchools = Districtsoup.find('div', class_="school_list split_columns")
    try:
        schools = listofSchools.find_all('a') 
    except:
        print("An exception occurred with link ", link)

    for i in range (len(schools)): #iterate through all the schools in each district
        schoolnames.append(schools[i].text)
        schoollinks.append(schools[i]['href'])
 
```

```python
teacherlink = []
teachername = []

for schoollink in schoollinks:
    Schoolpage = requests.get(schoollink, headers={'User-Agent': 'Mozilla/5.0'})
    Schoolsoup = BeautifulSoup(Schoolpage.content, 'html.parser')
    try:
        listofTeachers = Schoolsoup.find_all('div', class_ = "filter_target")
    except:
        print("An exception occurred with school link ", schoollink)
    try:
        teacher = listofTeachers[0].find_all('a') 
    except:
        print("Zero teacher ain't working with school link ", schoollink)
    for i in range (len(teacher)):
        teacherlink.append(teacher[i]['href'])
        teachername.append(teacher[i].text)
```

```python
teacherlinkFirst = teacherlink[15000:]
```

```python
#Trial with dictionary:
trackah = 0
keyList = ["Name:", "School Name:", "District Name:", "Base Salary:", "Other Salary:","Total Final Salary:", 
           "Insurance:", "Mandatory Benefits:", "Certificated FTE:", "Certificated Experience:", "Academic Credits:", 
           "Status:", "Present Age/Sex:", "In-Service Credits:", "ExcessCredits:", "Highest Degree:"] 

listofdicts = []      
#base, other, total, insurance, benefits, fte, experience, academiccredits, status, presentagesex, inservice, excess, degree = ([] for i in range(13)) 

for teacher in teacherlinkFirst:
    trackah = trackah + 1
    if (trackah % 50 == 0):
        print(trackah)
    dict = {} 
  
    for i in keyList: 
        dict[i] = None

    Teacherpage = requests.get(teacher, headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(Teacherpage.content, 'html.parser')
    try:
        teachinfo = soup.find_all('div', class_="data-card assignment even")
        locationinfo = teachinfo[0].find_all('div', class_ = "location")
        schoolname = locationinfo[0].text.split(" in ")[0]
        districtname = locationinfo[0].text.split(" in ")[1]
        dict["District Name:"] = districtname.replace('\n','')
        dict["School Name:"] = schoolname.replace('\n','')
        
    except:
        dict["District Name:"] = "N/A"
        dict["School Name:"] = "N/A"
    
    try:
        nameofperson = soup.find_all('header', class_ = "entry-header")
        name = nameofperson[0].find_all('h1', class_ = "entry-title")
        realname = name[0].text
        dict["Name:"] = realname
        
    except:
        dict["Name:"] = "N/A"
        

    try:
        data = soup.find_all('div', class_="data-card general") #All the years 
        currentyear = data[0]
    except:
        print("There is no current year for ", teacher)
    
    
    for strong_tag in currentyear.find_all(class_ = "label"):
        for key in dict:
            if (strong_tag.text == key):
                dict[key] = strong_tag.next_sibling.replace('\n','')
    #print(dict)
    listofdicts.append(dict)
        #print("label: ", strong_tag.text)
        #print("value: ", strong_tag.next_sibling)
        
```

```python

dataframe = pd.DataFrame(listofdicts)  
    
# Print the data  
dataframe.head(10)
```
